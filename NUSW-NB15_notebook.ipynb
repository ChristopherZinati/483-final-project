```python
import pandas as pd
from pathlib import Path

file_paths = [
    Path("C:/Users/maxpd/Desktop/483 ML/Intrusion Detection/483-final-project-datasets/UNSW-NB data/UNSW-NB15_1.csv"),
    Path("C:/Users/maxpd/Desktop/483 ML/Intrusion Detection/483-final-project-datasets/UNSW-NB data/UNSW-NB15_2.csv"),
    Path("C:/Users/maxpd/Desktop/483 ML/Intrusion Detection/483-final-project-datasets/UNSW-NB data/UNSW-NB15_3.csv"),
    Path("C:/Users/maxpd/Desktop/483 ML/Intrusion Detection/483-final-project-datasets/UNSW-NB data/UNSW-NB15_4.csv"),
]
balanced_dfs = []

for path in file_paths:
    df = pd.read_csv(path, header=None)

    label_col = df.iloc[:, -1]
    positives = df[label_col == 1]
    negatives = df[label_col == 0]

    negatives_sampled = negatives.sample(n=len(positives), random_state=42)

    balanced = pd.concat([positives, negatives_sampled], ignore_index=True).sample(frac=1, random_state=42)
    balanced_dfs.append(balanced)

final_balanced_df = pd.concat(balanced_dfs, ignore_index=True)
final_balanced_df.to_csv("balanced_UNSW_NB15.csv", index=False, header=False)
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 29)
```

```python
#Building a pipeline for transformations and modeling
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

categorical_cols = X_train.select_dtypes(include = 'object').columns.tolist()
numerical_columns = [item for item in X_train.columns if item not in categorical_cols]

preprocessor = ColumnTransformer(transformers = [('object', OneHotEncoder(handle_unknown= 'ignore', sparse_output = False), categorical_cols),
                                                 ('numbers', StandardScaler(), numerical_columns)])
pipeline_RFC = Pipeline([
    ('preprocessor' , preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=12,
                                          max_depth=12,
                                          min_samples_split=10,
                                          min_samples_leaf=5,
                                          random_state=42
                                         ))
])
```

```python
from sklearn.metrics import accuracy_score
NUSWModel1 = RandomForestClassifier(n_estimators=12,
                                          max_depth=12,
                                          min_samples_split=10,
                                          min_samples_leaf=5,
                                          random_state=42)
NUSWModel1.fit(preprocessor.fit_transform(X_train), y_train)
y_pred = NUSWModel1.predict(preprocessor.transform(X_test))
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy = ', accuracy)
encoded_feature_names = preprocessor.get_feature_names_out()
feature_importance = pd.Series(NUSWModel1.feature_importances_, index=encoded_feature_names)
feature_importance = feature_importance.sort_values(ascending=True)
print(feature_importance)
```

```python
feature_info = pd.DataFrame({
    'correlation': correlations_with_label,
    'importance': feature_importance
})
low_correlation_thresh = 0.10
low_importance_thresh = 0.005

to_drop = feature_info[
    (feature_info['correlation'].abs() < low_correlation_thresh) &
    (feature_info['importance'] < low_importance_thresh)
].index.tolist()

X_train_reduced = X_train.drop(columns=to_drop)
X_test_reduced = X_test.drop(columns=to_drop)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(transformers = [
                                                 ('numbers', StandardScaler(), numerical_columns)])
pipeline_KNN = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', KNeighborsClassifier())
])
k_vals = []
error_rates = []

for i in range(2, 40):
    pipeline_KNN.set_params(classifier__n_neighbors = i)
    pipeline_KNN.fit(X_train_reduced, y_train)
    error_rates.append(
        1 - accuracy_score(y_test, pipeline_KNN.predict(X_test_reduced))
    )
    k_vals.append(i)

plt.figure()
plt.title("Elbow Plot for k Values", fontsize=16)
plt.plot(k_vals, error_rates, linestyle='-')
plt.xlabel("number of neighbors")
plt.ylabel("error rate")
plt.show()
```
